### Data Prep
We'll use your provided CSV datasets. The objective is to format the text data for instruction-tuned language model fine-tuning. This usually involves creating a conversation-like structure where the model learns to respond to specific instructions based on the input text.

This cell loads the CSV files from the Hugging Face dataset.

```python
%%capture
!pip install pandas

import pandas as pd

# Direct URLs to the CSV files on Hugging Face
guidelines_url = "https://huggingface.co/datasets/gauravxthakur/PICURD/raw/main/guidelines.csv"
hadlock_url = "https://huggingface.co/datasets/gauravxthakur/PICURD/raw/main/hadlock.csv"
synthetic_reports_url = "https://huggingface.co/datasets/gauravxthakur/PICURD/raw/main/synthetic_fetalUS_reports.csv"

# Download and load guidelines.csv
guidelines_df = pd.read_csv(guidelines_url)

# Download and load hadlock.csv
hadlock_df = pd.read_csv(hadlock_url)

# Download and load synthetic_fetalUS_reports.csv
synthetic_reports_df = pd.read_csv(synthetic_reports_url)

print("Datasets loaded successfully:")
print(f"guidelines_df shape: {guidelines_df.shape}")
print(f"hadlock_df shape: {hadlock_df.shape}")
print(f"synthetic_reports_df shape: {synthetic_reports_df.shape}")
```

This cell inspects the first few rows of each DataFrame to understand their structure and content.

```python
print("\n--- guidelines_df.head() ---")
display(guidelines_df.head())

print("\n--- hadlock_df.head() ---")
display(hadlock_df.head())

print("\n--- synthetic_reports_df.head() ---")
display(synthetic_reports_df.head())
```

This cell creates functions to format the CSV data into a conversation-like structure, suitable for instruction-tuned language models. This allows the model to learn to summarize or extract information from the texts. Each function defines an instruction to frame the task, with the content from the data serving as input for the model to process.

```python
from datasets import Dataset, concatenate_datasets # Import concatenate_datasets

def format_guideline_sample(sample):
    instruction = "Provide information about the following medical guideline:"
    user_input = sample['TEXT']
    assistant_response = sample['TEXT']

    messages = [
        {"role": "user", "content": instruction + "\n" + user_input},
        {"role": "assistant", "content": assistant_response}
    ]
    return {"messages": messages}

def format_hadlock_sample(sample):
    instruction = "Provide details on the Hadlock measurement equation:"
    user_input = f"For BPD: {sample['data__bpd_mean']} mm, Gestational Age (GA): {sample['data__gestational_age_weeks']} weeks."
    # Hardcoding a common Hadlock BPD equation since it's not available in the dataframe.
    # A common Hadlock BPD equation (example) is: GA = 8.16 + 0.94(BPD) - 0.005(BPD^2)
    assistant_response = "The Hadlock BPD equation (simplified) is often approximated as: Gestational Age (weeks) = 8.16 + 0.94 * BPD - 0.005 * BPD^2"

    messages = [
        {"role": "user", "content": instruction + "\n" + user_input},
        {"role": "assistant", "content": assistant_response}
    ]
    return {"messages": messages}

def format_report_sample(sample):
    instruction = "Generate a fetal ultrasound report based on the following measurements:"
    head_circumference = sample['Head Circumference']
    lmp_ga = sample['LMP GA']
    scenario_type = sample['Scenario Type']
    report_text = sample['Report']

    user_input = f"Head Circumference: {head_circumference} mm, LMP GA: {lmp_ga}, Scenario Type: {scenario_type}."

    messages = [
        {"role": "user", "content": instruction + "\n" + user_input},
        {"role": "assistant", "content": report_text}
    ]
    return {"messages": messages}

# Create Hugging Face Dataset objects from pandas DataFrames
guidelines_hf_dataset = Dataset.from_pandas(guidelines_df)
hadlock_hf_dataset = Dataset.from_pandas(hadlock_df)
synthetic_reports_hf_dataset = Dataset.from_pandas(synthetic_reports_df)

# Format each dataset
formatted_guidelines_dataset = guidelines_hf_dataset.map(format_guideline_sample, remove_columns=guidelines_hf_dataset.column_names)
formatted_hadlock_dataset = hadlock_hf_dataset.map(format_hadlock_sample, remove_columns=hadlock_hf_dataset.column_names)
formatted_reports_dataset = synthetic_reports_hf_dataset.map(format_report_sample, remove_columns=synthetic_reports_hf_dataset.column_names)

# Combine all formatted datasets for training
formatted_dataset = concatenate_datasets([formatted_guidelines_dataset, formatted_hadlock_dataset, formatted_reports_dataset]) # Corrected concatenation

print("Formatted dataset example (first entry):")
display(formatted_dataset[0])
print(f"Total samples in formatted_dataset: {len(formatted_dataset)}")
```

This cell applies the Gemma 4 instruction chat template to the base model.

```python
from unsloth import get_chat_template

processor = get_chat_template(
    processor,
    "gemma-4", # For Gemma 4 models
)
```

Before fine-tuning, the base model's performance is evaluated. Strong results are not expected, as it has not encountered this chat template before.

```python
instruction = "Provide information about the following medical guideline:"
# Use a sample from your guidelines data for inference demonstration
example_guideline_data = guidelines_df.iloc[0]
# Corrected to use 'TEXT' column
input_text = example_guideline_data['TEXT']

messages = [
    {"role": "user", "content": instruction + "\\n" + input_text},
]

# Note: Gemma-4 uses a processor, not a tokenizer directly for chat templating with multimodal inputs
input_text_processed = processor.apply_chat_template(messages, add_generation_prompt = True)
inputs = processor(
    text=input_text_processed,
    return_tensors = "pt",
).to("cuda")

from transformers import TextStreamer

text_streamer = TextStreamer(processor.tokenizer, skip_prompt = True) # Use processor.tokenizer for streaming
result = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 256, # Increased max_new_tokens for potentially longer outputs
                        use_cache = True, temperature = 1.0, top_p = 0.95, top_k = 64)
```

The initial inference results show that the base model does not yet follow instructions effectively.

<a name="Train"></a>
### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support `DPOTrainer` and `GRPOTrainer` for reinforcement learning!

We use our new `UnslothVisionDataCollator` which will help in our vision finetuning setup.

```python
from unsloth.trainer import UnslothVisionDataCollator # Import for vision-capable models
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model = model,
    tokenizer = processor.tokenizer, # Use processor's tokenizer for SFTTrainer
    train_dataset = formatted_dataset,
    # dataset_text_field = "messages", # Removed to pass full dictionary samples to data_collator
    max_seq_length = 2048,
    data_collator = UnslothVisionDataCollator(model, processor), # For vision models
    args = SFTConfig(
        per_device_train_batch_size = 1,
        gradient_accumulation_steps = 4,
        max_grad_norm = 0.3,
        warmup_ratio = 0.03,
        num_train_epochs = 1,
        learning_rate = 2e-4,
        logging_steps = 1,
        save_strategy = "steps",
        optim = "adamw_8bit",
        weight_decay = 0.001,
        lr_scheduler_type = "cosine",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # For Weights and Biases or others
        # Vision-specific arguments for SFTTrainer
        remove_unused_columns = False,
        dataset_kwargs = {"skip_prepare_dataset": True},
        max_length = 2048,
    )
)
```

```python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
trainer_stats = trainer.train()
```

```python
# @title Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```

<a name="Inference"></a>
### Inference
Let's run the model! You can modify the instruction and input—just leave the output blank.

We'll use the best hyperparameters for inference on Gemma: `top_p=0.95`, `top_k=64`, and `temperature=1.0`.

```python
instruction = "Provide details on the Hadlock measurement equation:"
# Use a sample from your hadlock data for inference demonstration
example_hadlock_data = hadlock_df.iloc[0]
# Corrected column names based on hadlock_df structure
input_text = f"For BPD: {example_hadlock_data['data__bpd_mean']} mm, Gestational Age (GA): {example_hadlock_data['data__gestational_age_weeks']} weeks."

messages = [
    {"role": "user", "content": instruction + "\\n" + input_text},
]

input_text_processed = processor.apply_chat_template(messages, add_generation_prompt = True)
inputs = processor(
    text=input_text_processed,
    return_tensors = "pt",
).to("cuda")

from transformers import TextStreamer

text_streamer = TextStreamer(processor.tokenizer, skip_prompt = True)
result = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 256, # Increased max_new_tokens
                        use_cache = True, temperature = 1.0, top_p = 0.95, top_k = 64)
```
